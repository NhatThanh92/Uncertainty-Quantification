```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import random
import pandas as pd
import seaborn as sns
```

```python
axis_font = {'fontname':'Arial', 'size':'20'}
sns.set(rc = {'figure.figsize':(8,5)})
sns.set(font_scale = 1.5)
```

```python
###### P-box or Horsetail plot 
# Area between The upper and lower bounds of the CDF in the horsetail plot known as the probability box or p-box. 
# The p-box represents a range of CDFs that can represent the system.
# Distributional p-boxes, When a probability distribution is known to have a particular shape 
# but its parameters can only be specified imprecisely as intervals,
# the result is called a distributional p-box, or sometimes a parametric p-box 


# ####     FOR C10

r = np.arange(0.02,0.08,1e-4)
min = [1 for _ in range(len(r))]; max = [0 for _ in range(len(r))]
plt.figure(figsize=(9,7))

for _ in range(100):
    a= random.uniform(0.04,0.05) # mean
    b= random.uniform(0.005,0.01)  # std
    #a= random.uniform(0.04,0.05) # mean
    #b= random.uniform(0.003,0.01)  # std
    plt.plot(r, norm.cdf(r,a,b),'r--',lw=1,alpha=0.4,scalex=True, scaley=True) # remove ('k--',lw=1,alpha=0.4) => get color plot
    
#plt.title('Figure 1: Horsetail Plot of 100 Potential CDFs')
plt.ylabel('CDF',fontsize=16)
plt.xlabel('C10',fontsize=16)
plt.legend(['randomized CDFs'],fontsize=16)
plt.show()
```

```python
r = np.arange(0.02,0.08,1e-4)
min = [1 for _ in range(len(r))]; max = [0 for _ in range(len(r))]
plt.figure(figsize=(9,6))

for _ in range(100):
    a= random.uniform(0.04,0.05) # mean
    b= random.uniform(0.005,0.01)  # std
    #a= random.uniform(0.04,0.05) # mean
    #b= random.uniform(0.003,0.01)  # std
    Fx = norm.cdf(r,a,b)
    #plt.plot(r, Fx)
    for i in range(len(r)):
        if Fx[i]<min[i]:
            min[i] = Fx[i]
        elif Fx[i]>max[i]:
            max[i] = Fx[i]
 ########
line_color = (0, 0, 0, 0.5)           
# INTERPOLATION  for MAX          
xs = np.sort(r)
ys = np.array(max)[np.argsort(r)]
# x coordinate
y95 = 0.95
# interpolated y coordinate
xma = np.interp(y95, ys, xs)
plt.plot(xma,y95, marker="o")
plt.plot([xma,xma,0.02],[0.02,y95,y95],'C2--', color=line_color)
#############
# INTERPOLATION  for MIN         
xsi = np.sort(r)
ysi = np.array(min)[np.argsort(r)]
# x coordinate

# interpolated y coordinate
xmi = np.interp(y95, ysi, xsi)
plt.plot(xmi,y95, marker="o")
plt.plot([xmi,xmi,0.02],[0,y95,y95],'C2--', color=line_color)

##
plt.text(0.012,0.93,'${0.95}$', fontsize=11, color='red')
plt.plot(r,max,'C1--',label='max',lw=3, color='blue')
plt.plot(r,min,'C2--',label='min',lw=3)
plt.title('Figure 2: Horsetail Plot of 100 Potential CDFs')
plt.text(0.035,0.8,'$P_{upper}$(x)', fontsize=13, color='blue')
plt.text(0.041,0.025,'$P_{lower}$(x)', fontsize=13, color='green' )
#plt.plot(r, norm.cdf(r,0.045,0.008),label='exact CDF', color='red')
plt.ylabel('CDF')
plt.xlabel('C10')
plt.legend()
plt.show()
```
